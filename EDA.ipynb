```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 2. Load Dataset
df = pd.read_csv("/content/heart_cleveland_upload.csv")  # Adjust filename if needed

# 3. Basic Info
print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
```

```python

# 4. Summary Statistics
print("\nSummary Statistics:\n", df.describe())

```

```python
# 5. Histograms for Numeric Features
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
df[numeric_cols].hist(bins=15, figsize=(16, 12), layout=(len(numeric_cols)//3+1, 3))
plt.suptitle("Histograms of Numeric Features")
plt.tight_layout()
plt.show()
```

```python

# 6. Boxplots for Outlier Detection
for col in numeric_cols:
    plt.figure()
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python

# 7. Correlation Matrix
plt.figure(figsize=(12, 10))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Correlation Heatmap")
plt.show()
```
